```python
def validate_path(path):
    if not os.path.exists(path):
        raise FileNotFoundError(f"File not found: {path}")
    
def validate_data(data : Dict):
    validate_path(data['robots'])
    validate_path(data['objects'])
    validate_path(data['scene'])
    validate_path(data['obstacles'])

    print("SUCCESS")
```

```python
shelf_stats = {}
shelf_stats['2'] = {}
shelf_stats['3'] = {}
shelf_stats['4'] = {}
for key in shelf_stats.keys():
    shelf_stats[key]['travel_time'] = []
    shelf_stats[key]['path_length'] = []
    shelf_stats[key]['total_rrt_time'] = []
    shelf_stats[key]['total_coll_time'] = []
    shelf_stats[key]['total_nn_time'] = []
```

```python
def clear_dir(directory):
    for filename in os.listdir(directory):
            file_path = os.path.join(directory, filename)
            if os.path.isfile(file_path):
                os.remove(file_path)
                print(f"Deleted: {file_path}")

def get_num_obj(directory):
     with open(directory, 'r') as f:
        data = json.load(f)
        return len(data['objects'])
```

```python
directory = "/home/alex/multitask/24-data-gen/log"

seeds = [879, 1, 325, 16, 615]
for folder in tqdm(folders[10:30]):
    num_objects = get_num_obj(f'/home/alex/multitask/24-data-gen/in/shelf/{folder}/obj_path.json')    
    for seed in seeds:
        clear_dir(directory)

        args = { 'robots':    f'in/shelf/{folder}/robot_path.json',
                  'objects':   f'in/shelf/{folder}/obj_path.json',
                  'scene':     f'in/shelf/{folder}/scene.g',
                  'obstacles': f'in/shelf/{folder}/obstacles_file.json'}

        command = [
            './x.exe',
            '-pnp',             'true',
            '-mode',            'random_search',
            '-seed',            str(seed),
            '-robot_path',      args['robots'],
            '-obj_path',        args['objects'],
            '-attempt_komo',    'false',
            '-display',         'false',
            '-export_images',   'false',
            '-scene_path',      args['scene'],
            '-obstacle_path',   args['obstacles']
        ]
        validate_data(args)
        subprocess.run(command)

        travel_time = 0.0
        path_length = 0.0
        total_rrt_time = 0.0
        total_coll_time = 0.0
        total_nn_time = 0.0
        
        for filename in os.listdir('/home/alex/multitask/24-data-gen/log'):
            if filename.endswith('.json'):
                try:
                    with open('/home/alex/multitask/24-data-gen/log/' + filename, 'r') as f:
                        data = json.load(f)

                        if data['travel_time'] != 0:
                            travel_time     += data['travel_time']
                            path_length     += data['path_length']
                            total_rrt_time  += data['total_rrt_time']
                            total_coll_time += data['total_coll_time']
                            total_nn_time   += data['total_nn_time']

                        print(f"Содержимое файла {filename}:")
                        print(json.dumps(data, indent=4))
                    
                except Exception as e:
                    print(f"Ошибка при обработке файла {filename}: {e}\n")

        shelf_stats[str(num_objects)]['travel_time'].append(travel_time)
        shelf_stats[str(num_objects)]['path_length'].append(path_length)
        shelf_stats[str(num_objects)]['total_rrt_time'].append(total_rrt_time)
        shelf_stats[str(num_objects)]['total_coll_time'].append(total_coll_time)
        shelf_stats[str(num_objects)]['total_nn_time'].append(total_nn_time)        
```

```python
file_path = "/home/alex/multitask/24-data-gen/stats.json"
with open(file_path, 'w', encoding='utf-8') as json_file:
    json.dump(shelf_stats, json_file, ensure_ascii=False, indent=4)
```
